```python
if index_name not in pc.list_indexes():  # Correctly checks if the index exists
    pc.create_index(
        name=index_name,
        dimension=384,  # Dimension of the dense vector
        metric='dotproduct'  # Ensure the metric is appropriate for your use case
    )
else:
    print(f"Index '{index_name}' already exists.")
```

```python
index = pc.Index(index_name)
```

```python
from langchain.embeddings import HuggingFaceEmbeddings
```

```python
embedding_model = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")  # A smaller transformer model
```

```python
documents = [
    "In 2023, I visited Paris",
    "In 2022, I visited New York",
    "In 2021, I visited New Orleans"
]
```

```python
document_embeddings = embedding_model.embed_documents(documents)
```

```python
ids = ["doc1", "doc2", "doc3"]
```

```python
# Prepare the data to upsert
# Assuming each document has an associated unique ID
ids = ["doc1", "doc2", "doc3"]

# Create the data to upsert (it can be a list of dictionaries with 'id', 'vector', and 'metadata')
upsert_data = [
    {"id": doc_id, "values": embedding, "metadata": {"text": doc_text}}
    for doc_id, embedding, doc_text in zip(ids, document_embeddings, documents)
]

# Upsert into the Pinecone index
index.upsert(vectors=upsert_data)
```

```python
# Example query document
query = "I want to travel to Paris"

# Get the embedding for the query
query_embedding = embedding_model.embed_documents([query])[0]

# Query Pinecone for the most similar document
query_result = index.query(
    vector=query_embedding,
    top_k=1,  # Number of similar documents to retrieve
    include_metadata=True  # To get metadata (e.g., original document text)
)

# Print the results
print(query_result)
```
